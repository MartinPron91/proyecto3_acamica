```python
### Se importan las librerias a utilizar
import pandas as pd
import plotly.express as px
```

```python

aqm_rosa=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/LAI/Datos UMMCA viento direccion.csv")
aqm_rosa
```

```python
###Se transforma la variable con la velocidad del viento en una variable categórica. Se definen 5 categorías con cuantiles###
valor_1= 0
valor_2= aqm_rosa['WS (m/s    )'].quantile(0.2)
valor_3= aqm_rosa['WS (m/s    )'].quantile(0.4)
valor_4= aqm_rosa['WS (m/s    )'].quantile(0.6)
valor_5= aqm_rosa['WS (m/s    )'].quantile(0.8)
valor_6= aqm_rosa['WS (m/s    )'].max()

valores=[valor_1,valor_2, valor_3, valor_4,valor_5, valor_6]
categoria_1 = "0 - " +str(valor_2)
categoria_2 = str(valor_2)+" - "+str(valor_3)
categoria_3 = str(valor_3)+" - "+str(valor_4)
categoria_4 = str(valor_4)+" - "+str(valor_5)
categoria_5 = ">" +str(valor_5)


categorías=[categoria_1, categoria_2, categoria_3, categoria_4, categoria_5] 
aqm_rosa["categoria_velocidades"]=pd.cut(x=aqm_rosa['WS (m/s    )'],bins=valores,labels=categorías)
aqm_rosa["categoria_velocidades"].unique()
```

```python
### Se corrabora que se haya realizado bien la transformación. Se muestran los valores faltantes en caso de fallo ###
nulos = pd.isnull(aqm_rosa["categoria_velocidades"]) 
aqm_rosa[nulos]
```

```python
#aqm_rosa["categoria_velocidades"].value_counts().reset_index(name="numero de repeticiones")
```

```python
### Definimos 5 dataframes, uno por cada categoría de velocidades
df_1 = aqm_rosa.loc[aqm_rosa['categoria_velocidades']==categoria_1]
df_2 = aqm_rosa.loc[aqm_rosa['categoria_velocidades']==categoria_2]
df_3 = aqm_rosa.loc[aqm_rosa['categoria_velocidades']==categoria_3]
df_4 = aqm_rosa.loc[aqm_rosa['categoria_velocidades']==categoria_4]
df_5 = aqm_rosa.loc[aqm_rosa['categoria_velocidades']==categoria_5]
```

```python
### Definimos un diccionario con el critorio de ordenación de las coordenadas polares
dic = {"N":1, "NNE":2, "NE":3,"ENE":4, "E":5,"ESE":6,"SE":7,"SSE":8,"S":9,"SSO":10,"SO":11,"OSO":12,"O":13,"ONO":14,"NO":15,"NNO":16}
```

```python
### Para el Data Frame 1 Se Agrupan los datos por el recuento de valores en cada dirección y se dibide ese valor por el recuento total, obteniendose la frecuencia. Se ordena de N a NNO
df_1= df_1[['WS (m/s    )','WD-1 (deg    )']].groupby('WD-1 (deg    )').agg(pd.Series.count)
df_1["WS (m/s    )"] = df_1["WS (m/s    )"]/len(aqm_rosa.index)*100
df_1.reset_index(inplace=True)
df_1.columns = ["Dirección","Frecuencia"]
df_1["Categoría Velocidad"]= categoria_1
df_1["criterio"]=df_1.Dirección.map(dic)
df_1.sort_values(by=["criterio"],inplace=True)
df_1 = df_1.reset_index().drop(columns=["index","criterio"])
```

```python
### Se repite lo mismo para los otro 4 dataframes.
df_2= df_2[['WS (m/s    )','WD-1 (deg    )']].groupby('WD-1 (deg    )').agg(pd.Series.count)
df_2["WS (m/s    )"] = df_2["WS (m/s    )"]/len(aqm_rosa.index)*100
df_2.reset_index(inplace=True)
df_2.columns = ["Dirección","Frecuencia"]
df_2["Categoría Velocidad"]= categoria_2
df_2["criterio"]=df_2.Dirección.map(dic)
df_2.sort_values(by=["criterio"],inplace=True)
df_2 = df_2.reset_index().drop(columns=["index","criterio"])
```

```python
df_3= df_3[['WS (m/s    )','WD-1 (deg    )']].groupby('WD-1 (deg    )').agg(pd.Series.count)
df_3["WS (m/s    )"] = df_3["WS (m/s    )"]/len(aqm_rosa.index)*100
df_3.reset_index(inplace=True)
df_3.columns = ["Dirección","Frecuencia"]
df_3["Categoría Velocidad"]= categoria_3
df_3["criterio"]=df_3.Dirección.map(dic)
df_3.sort_values(by=["criterio"],inplace=True)
df_3 = df_3.reset_index().drop(columns=["index","criterio"])
```

```python
df_4= df_4[['WS (m/s    )','WD-1 (deg    )']].groupby('WD-1 (deg    )').agg(pd.Series.count)
df_4["WS (m/s    )"] = df_4["WS (m/s    )"]/len(aqm_rosa.index)*100
df_4.reset_index(inplace=True)
df_4.columns = ["Dirección","Frecuencia"]
df_4["Categoría Velocidad"]= categoria_4
df_4["criterio"]=df_4.Dirección.map(dic)
df_4.sort_values(by=["criterio"],inplace=True)
df_4 = df_4.reset_index().drop(columns=["index","criterio"])
```

```python
df_5= df_5[['WS (m/s    )','WD-1 (deg    )']].groupby('WD-1 (deg    )').agg(pd.Series.count)
df_5["WS (m/s    )"] = df_5["WS (m/s    )"]/len(aqm_rosa.index)*100
df_5.reset_index(inplace=True)
df_5.columns = ["Dirección","Frecuencia"]
df_5["Categoría Velocidad"]= categoria_5
df_5["criterio"]=df_5.Dirección.map(dic)
df_5.sort_values(by=["criterio"],inplace=True)
df_5 = df_5.reset_index().drop(columns=["index","criterio"])
```

```python
### Se concatenan los dataframe ordenados del de menor categoría de velocidades al de mayor categoría
df= pd.DataFrame()
df = pd.concat([df_1, df_2, df_3, df_4, df_5], axis=0)
df.reset_index(inplace=True)
df.drop(columns="index",inplace=True)
```

```python
fecha_ini = aqm_rosa["Date"].unique()[0]
fecha_fin = aqm_rosa["Date"].unique()[1]
```

```python
### Se realiza la rosa de los vientos mediante plotly

fig = px.bar_polar(df, r="Frecuencia", theta="Dirección", color="Categoría Velocidad", template="plotly_dark",
            color_discrete_sequence= px.colors.sequential.Plasma_r)
fig.update_layout(
    title={
        #Cadena de Custodia y Fecha
        'text': fecha_ini+" a "+fecha_fin, 
        'y':0.05,
        'x':0.467,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.show()
```
